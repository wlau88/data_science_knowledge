#EDA (Exploratory Data Analysis)

####Exploratory data analysis is a first crucial step to building predictive models from your data. EDA allows you to confirm or invalidate some of the assumptions you are making about your data and understand relationships between your variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
import scipy.stats as scs
```

```python
df = pd.read_csv('data/201402_trip_data.csv', parse_dates=['start_date', 'end_date'])
```

```python
df = df.set_index(['start_date'])
```

####Engineer some extra features to look at (month, day of week, date, hour)

```python
df['month'] = df.index.month
```

```python
df['dayofweek'] = df.index.dayofweek
```

```python
df['date'] = df.index.date
```

```python
df['hour'] = df.index.hour
```

```python
df.columns
```

####Group the bike rides by month and count up the number of users per month

```python
dmonth = df.groupby('month')['trip_id'].count()
```

```python
x = df.month.unique()
y = []
for i in df.month.unique():
    y.append(dmonth[i])

y = np.array(y)
plt.xlabel('month')
plt.ylabel('rider count')
plt.bar(x,y)
plt.show()
```

####Exploring stations with low usage

```python
dfstations = pd.read_csv('data/201402_station_data.csv')
dfstations.head()
```

```python
df_usage = pd.DataFrame(df.groupby('start_station')['trip_id'].count()).reset_index()
df_usage.rename(columns={'start_station': 'name'}, inplace=True)
dfstations_merged = dfstations.merge(df_usage, how='left', on='name')
dfstations_merged['usage'] = dfstations_merged['trip_id'] / dfstations_merged['dockcount']
dfstations_merged['norm_usage'] = dfstations_merged['usage'] / dfstations_merged['usage'].sum()
dfstations_merged.head()
```

```python
sizes = dfstations_merged['usage']
plt.scatter(dfstations_merged['long'], dfstations_merged['lat'], s=sizes, marker='o', c=sizes)
plt.show()
```

#Linear Regression intro

```python
import statsmodels.api as sm
prestige = sm.datasets.get_rdataset("Duncan", "car", cache=True).data
y = prestige['prestige']
x = prestige[['income', 'education']].astype(float)
from pandas.tools.plotting import scatter_matrix
y = pd.DataFrame(y)
x = x.reset_index()
y = y.reset_index()
m = x.merge(y, how='inner').set_index('index')
scatter_matrix(m)
plt.show()
```

```python
m.boxplot()
plt.show()
```

The beta coefficients of a linear regression model can be calculated by solving the normal equation:

   <div align="center">
      <img height="30" src="images/normal_equation.png">
   </div>

Below is an implementation of that:

```python
def solv_eq(x, y):
    return np.dot(np.dot(np.linalg.inv(np.dot(x.T, x)),x.T),y)

x = m[['income', 'education']]
y = m[['prestige']]
x = np.array(x)
y = np.array(y)

B = solv_eq(x,y)
```

```python
B
```

```python
y_hat = np.dot(x,B)
TSS = ((y - np.mean(y))**2).sum()
RSS = ((y - y_hat)**2).sum()

print RSS, TSS
```

```python
1-RSS/float(TSS)
```

```python
import statsmodels.api as sms
model = sms.OLS(y, x).fit()
summary = model.summary()
```

```python
summary
# The beta coefficients match with our calculated results, 
# the R^2 is very good, p-value is small, F-statistic is large (which means
# at least one of the predictors is useful at predicting responses,
# it's similar to t-test in regression output)
```

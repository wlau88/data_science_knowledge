##K-means implementation and example - Iris dataset

```python
import numpy as np
from sklearn.datasets import load_iris
import random
data = load_iris()
import copy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = data.data
```

```python
X.shape
```

```python
from kmeans import *
```

###Checking SSE, Silhouette score and visualizing feature space

```python
plot_k_sse(X)
```

```python
plot_k_silhouette(X)
```

```python
plot_all_2d(X, np.array(data.feature_names), k=5)
```

```python
from sklearn import metrics
```

```python
new_X, labels = turn_clusters_into_labels(k_means(X))
metrics.silhouette_score(new_X, labels, metric='euclidean')
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=5, random_state=1).fit(X)
labels = kmeans_model.labels_
metrics.silhouette_score(X, labels, metric='euclidean')
```

####Silhouette score is similar to sklearn's implementation!

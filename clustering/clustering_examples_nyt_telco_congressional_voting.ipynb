#Clustering of NYT articles

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pdb
import random
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.pylab
from collections import Counter
```

```python
df = pd.read_pickle('data/articles.pkl')
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform(df['content'])
words = vectorizer.get_feature_names()
```

```python
len(words)
```

#K-means clustering

```python
kmeans = KMeans()
kmeans.fit(vectors)
```

```python
print "cluster centers:"
print kmeans.cluster_centers_
```

```python
top_centroids = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
print "top features for each cluster:"
for num, centroid in enumerate(top_centroids):
    print "%d: %s" % (num, ", ".join(words[i] for i in centroid))
```

#####There are clear themes in the cluster. Let's try with a fix max_features

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
vectors = vectorizer.fit_transform(df['content'])
words = vectorizer.get_feature_names()
kmeans = KMeans()
kmeans.fit(vectors)
top_centroids = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
print "top features for each cluster with 1000 max features:"
for num, centroid in enumerate(top_centroids):
    print "%d: %s" % (num, ", ".join(words[i] for i in centroid))
```

####Here are a sample of titles associated with each cluster

- politics: 0, 6
- sports: 3, 7
- international: 4
- business: 5
- arts: 2

```python
assigned_cluster = kmeans.transform(vectors).argmin(axis=1)
for i in range(kmeans.n_clusters):
    cluster = np.arange(0, vectors.shape[0])[assigned_cluster==i]
    sample_articles = np.random.choice(cluster, 3, replace=False)
    print "cluster %d:" % i
    for article in sample_articles:
        print "    %s" % df.ix[article]['headline']
```

###Set k to the number of sections in NYT

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(vectors)
assigned_cluster = kmeans.transform(vectors).argmin(axis=1)
print "top 2 topics for each cluster"
for i in range(kmeans.n_clusters):
    cluster = np.arange(0, vectors.shape[0])[assigned_cluster==i]
    topics = df.ix[cluster].dropna()['section_name']
    most_common = Counter(topics).most_common()
    print "Cluster %d: %s" % (i, most_common[0][0]),
    if len(most_common) > 1:
        print " %s" % (most_common[1][0]),
    print
```

#####The cluster classification is pretty good.

###Limit to just a subset of sections

```python
mask = np.logical_or(
       np.logical_or((df['section_name']=='Sports').values,
                     (df['section_name']=='Arts').values),
                     (df['section_name']=='Business Day').values)
three_articles_df = df[mask]
kmeans = KMeans(n_clusters=3)
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(three_articles_df['content'])
kmeans.fit(X)
assigned_cluster = kmeans.transform(X).argmin(axis=1)
print "top 2 topics for each cluster"
for i in range(kmeans.n_clusters):
    cluster = np.arange(0, X.shape[0])[assigned_cluster==i]
    topics = three_articles_df.ix[cluster].dropna()['section_name']
    most_common = Counter(topics).most_common()
    print "Cluster %d: %s" % (i, most_common[0][0]),
    if len(most_common) > 1:
        print " %s" % (most_common[1][0]),
    print
```

# Hierarchical Clustering

####Subset the data such that it contains at least one article from each section and at most around 200 total articles

```python
small_mask = np.zeros(len(df)).astype(bool)
indices = np.arange(len(df))
for category in df['section_name'].unique():
    category_mask = (df['section_name']==category).values
    new_index = np.random.choice(indices[category_mask])
    small_mask[new_index] = True
additional_indices = np.random.choice(indices[np.logical_not(small_mask)],
                                      100 - sum(small_mask),
                                      replace=False)
small_mask[additional_indices] = True
small_df = df.ix[small_mask]
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
small_X = vectorizer.fit_transform(small_df['content'])
small_features = vectorizer.get_feature_names()
```

```python
distxy = squareform(pdist(small_X.todense(), metric='cosine'))
```

```python
link = linkage(distxy, method='complete')
```

```python
dendro = dendrogram(link, color_threshold=1.5, leaf_font_size=9)
plt.show()
```

```python
dendro = dendrogram(link, color_threshold=1.5, leaf_font_size=9,
                    labels=small_df['headline'].values)

# fix spacing to better view dendrogram and the labels
plt.subplots_adjust(top=.99, bottom=0.5, left=0.05, right=0.99)
plt.savefig('big_dendrogram')
plt.show()
```

```python
# Explore different clusters on a per section basis.
for i, category in enumerate(['Arts', 'Sports', 'World']):
    cat_indices = df[df['section_name']==category].index
    indices = np.random.choice(cat_indices, 40)
    cat_df = df.ix[indices]
    vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
    cat_X = vectorizer.fit_transform(cat_df['content'])
    distxy = squareform(pdist(cat_X.todense(), metric='cosine'))
    ax = plt.subplot(1, 3, i + 1)
    dendro = dendrogram(linkage(distxy, method='complete'),
                        color_threshold=4,
                        leaf_font_size=4,
                        labels=small_df['headline'].values)
    ax.set_title(category)
plt.subplots_adjust(top=.95, bottom=0.5, left=0.05, right=0.99)
plt.savefig('section_clusters')
plt.show()
```

```python
# words from the articles.
plt.figure(figsize=(10,50))
distxy_words = squareform(pdist(small_X.T.todense(), metric='cosine'))
dendro = dendrogram(linkage(distxy_words, method='complete'),
                    color_threshold=1.5, leaf_font_size=3,
                    labels=small_features, leaf_rotation=0, orientation='left')
plt.subplots_adjust(top=.99, bottom=0.2, left=0.05, right=0.99)
plt.savefig('dendrogram_words_from_articles')
plt.show()
```

##Telco Dataset

```python
df_telco = pd.read_csv('data/churn.csv')
```

```python
df_telco.columns
```

```python
# Remove the features which aren't continuous or boolean
df_telco.pop('State')
df_telco.pop('Area Code')
df_telco.pop('Phone')
```

```python
# Convert the "no", "yes" values to booleans (True/False)
df_telco["Int'l Plan"] = df_telco["Int'l Plan"] == 'yes'
df_telco["VMail Plan"] = df_telco["VMail Plan"] == 'yes'
df_telco['Churn?'] = df_telco['Churn?'] == 'True.'
```

```python
X = df_telco.values
```

```python
colors = ["#000000","#FFFF00","#1CE6FF","#FF34FF","#FF4A46","#008941","#006FA6","#A30059",
      "#FFDBE5","#7A4900","#0000A6","#63FFAC","#B79762","#004D43","#8FB0FF","#997D87",
      "#5A0007","#809693","#FEFFE6","#1B4400","#4FC601","#3B5DFF","#4A3B53","#FF2F80",
      "#61615A","#BA0900","#6B7900","#00C2A0","#FFAA92","#FF90C9","#B903AA","#D16100",
      "#DDEFFF","#000035","#7B4F4B","#A1C299","#300018","#0AA6D8","#013349","#00846F",
      "#372101","#FFB500","#C2FFED","#A079BF","#CC0744","#C0B9B2","#C2FF99","#001E09",
      "#00489C","#6F0062","#0CBD66","#EEC3FF","#456D75","#B77B68","#7A87A1","#788D66",
      "#885578","#FAD09F","#FF8A9A","#D157A0","#BEC459","#456648","#0086ED","#886F4C"]
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
assigned_cluster = kmeans.transform(X).argmin(axis=1)
for i in range(kmeans.n_clusters):
    print "Cluster " + str(i+1)
    cluster = np.arange(0, X.shape[0])[assigned_cluster==i]
    
    heights = df_telco.ix[cluster, :].mean(axis=0).values
    
    plt.bar(np.arange(X.shape[1]), heights, 
            label='cluster %d' % (i+1), alpha=0.2, color=colors[i])
    plt.xticks(np.arange(X.shape[1]) + .5, [name for name in df_telco.columns], rotation=90)
    
plt.legend()
plt.show()    
```

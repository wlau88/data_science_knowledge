###Note on subset selection:

* [RSS](https://en.wikipedia.org/wiki/Residual_sum_of_squares) - metric linear regression trying to minimize
* [Mallow's C_p](https://en.wikipedia.org/wiki/Mallows%27s_Cp) - based on RSS, p, and variance of the error term (epsilon) in regression
* [AIC](https://en.wikipedia.org/wiki/Akaike_information_criterion) - based on L (maximum value of the likelihood function for the model estimated), and p (# of predictors) lower the better
* [BIC](https://en.wikipedia.org/wiki/Bayesian_information_criterion) - basically AIC but places more penalty on more variables
* [Adjusted R^2](https://en.wikipedia.org/wiki/Coefficient_of_determination) - R^2 but pays for more variables

```python
# Using RFE to evaluate subset selection (based on R^2)
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
y = pd.DataFrame(df.Balance)
```

```python
x_w_intercept.shape, y.shape
```

```python
x_w_intercept = sms.add_constant(x)
```

```python
r_squared_lst = []
adj_r_squared_lst = []
```

```python
import pdb
```

```python
n_features = len(x_w_intercept.columns)

for i,_ in enumerate(x_w_intercept.columns):
    i += 1
    lr = LinearRegression()
    selector = RFE(lr, n_features_to_select=i, step=1)
    selector.fit(x_w_intercept, y)
    
    parameters_selected = ', '.join(x_w_intercept.columns[selector.get_support()])
    print 'parameters selected: ', parameters_selected
    r_squared = selector.score(x_w_intercept,y)
    n_sample_size = x_w_intercept.shape[0]
    r_squared_lst.append(r_squared)
    # adjusted r_squared to take into account model complexity
    adj_r_squared = 1 - ((1 - r_squared)*(n_sample_size - 1))/(n_sample_size - i - 1)
    adj_r_squared_lst.append(adj_r_squared)
    print r_squared, adj_r_squared
    
    ### Model summary, Studentized residual plots
    print 1
    bool_index = np.arange(n_features)[selector.get_support()]
#     pdb.set_trace()
    features_subset = sms.add_constant(x_w_intercept.iloc[:,bool_index])
    target = y
    model = sms.OLS(target, features_subset).fit()
    print model.summary()
    student_res = model.outlier_test()['student_resid']
    fitted_y = model.fittedvalues
    plt.figure(figsize=(15,40))
    plt.subplot(n_features, 1, i, )
    plt.scatter(fitted_y, student_res, alpha=0.4)
    plt.title('parameters selected: ' + parameters_selected)
    plt.show()
```

```python
len(r_squared_lst), len(adj_r_squared_lst)
```

```python
plt.plot(np.arange(1,13), r_squared_lst, label='r_squared')
plt.plot(np.arange(1,13), adj_r_squared_lst, label='adj_r_squared')
plt.ylabel('metric')
plt.xlabel('no. of features')
plt.title('Feature importance')
plt.legend()
plt.show()
```

###Since the abundance of observations loaded at 0 are affecting the fit. One option would be to fit more than one model. In this case, our first "model" will predict 0 or non-zero balance based on just one feature. (Decision Trees can better model this..). The threshold that we've decided on are:

* Income < 50
* Limit < 3000
* Rating < 250


```python
df_nz = df[(df.Income > 50) | (df.Limit > 3150) | (df.Rating > 220)]
x_nz = df_nz.drop(['Balance', 'Rating'], axis=1)
y_nz = df_nz.Balance
y_nz.hist()
# The number of 0 balances are a lot less
```

```python
model_nz = sms.OLS(y_nz,x_nz).fit()
summary_nz = model_nz.summary()
summary_nz
# model to predict non-zero balance is good
```

```python
student_res_nz = model_nz.outlier_test()['student_resid']
fitted_y_nz = model_nz.fittedvalues
plt.scatter(fitted_y_nz, student_res_nz, alpha=0.4)
plt.show()
# Assumes more homoscedasticity
```

#Interactions, polynomials, transformations

```python
x_i = df_nz.Income
x_s = df_nz.Student
x_2 = df_nz[['Student', 'Income']]

y = df_nz.Balance
print x_i.shape, x_s.shape
```

###Model of Student, Income to predict Balance

###Model of Student, Income, Income * Student to predict Balance

```python
x_2['S*I'] = x_2['Student']*x_2['Income']
model_3 = sms.OLS(y, x_2).fit()
summary_3 = model_3.summary()
summary_3
# the model with interaction terms is better, the interaction term is significant
```

```python
x_2['fitted_y'] = model_3.fittedvalues
x_student = x_2[x_2['Student'] == 1]
x_nonstudent = x_2[x_2['Student'] == 0]

plt.plot(x_student['Income'], x_student['fitted_y'], label='student')
plt.plot(x_nonstudent['Income'], x_nonstudent['fitted_y'], label='non-student')
plt.ylabel('Balance')
plt.xlabel('Income')
plt.legend(loc=4)
plt.show()
# balance for student and non-student are different
```
